```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import pandas as pd
import numpy as np
import matplotlib

df = pd.read_csv("data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv")
df = df.drop('ID',axis=1)
df_index = pd.DatetimeIndex(df['Date'])
df = df[~((df_index.is_leap_year)&(df_index.month ==2)&(df_index.day == 29))]
df_index = pd.DatetimeIndex(df['Date'])
df['Month'] = df_index.month
df['Day'] = df_index.day

df_2015 = df[df_index >= "2015-01-01"]
df = df[df_index < "2015-01-01"]

df_max = df[df['Element']=='TMAX'].groupby(['Month','Day'])['Data_Value'].max()
df_max = df_max / 10

df_min = df[df['Element']=='TMIN'].groupby(['Month','Day'])['Data_Value'].min()
df_min = df_min / 10

df_2015_max = df_2015[df_2015['Element']=='TMAX'].groupby(['Month','Day'])['Data_Value'].max()
df_2015_max = df_2015_max / 10

df_2015_min = df_2015[df_2015['Element']=='TMIN'].groupby(['Month','Day'])['Data_Value'].min()
df_2015_min = df_2015_min / 10
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import matplotlib.dates as mdates
mpl.rcParams['figure.figsize'] = (16, 8)

max_broken = (df_2015_max > df_max).as_matrix()
# print(np.sum(max_broken))
min_broken = (df_2015_min < df_min).as_matrix()
# print(np.sum(min_broken))

xtime = np.arange('2015-01-01', '2016-01-01', dtype='datetime64[D]')
xtime = np.array(list(map(pd.to_datetime, xtime)))



plt.plot(xtime, df_min.as_matrix(), '-o', color='green', ms=2, alpha=0.75) 
plt.plot(xtime, df_max.as_matrix(), '-o', color='orange', ms=2, alpha=0.75) 

#2015 record temperatures
plt.plot(xtime[max_broken], df_2015_max[max_broken].as_matrix(), 'r^',
         xtime[min_broken], df_2015_min[min_broken].as_matrix(), 'bv', 
         ms=8, alpha=1.0)


plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.suptitle("2015 temperature records vs past 10 years (2005-2014) records (°C)", fontsize=16)
plt.title("Michigan", fontsize=14)
plt.legend(['Past Record Lows', 'Past Record Highs', "2015 record Highs", "2015 record Lows"], loc='center left', bbox_to_anchor=(1.0, 0.5), prop={'size':12})

plt.gca().fill_between(xtime, df_min, df_max, facecolor='grey', alpha=0.15)
```
